```python
df = pd.read_csv("../data/raw/insurance_data.csv")

print(df.head())
```

```python
claims_df = df[df["TotalClaims"] > 0]

print(claims_df.shape)
```

# Statistical Modeling

The objective of this section is to build predictive models for insurance claim severity.

Three algorithms are evaluated:

1. Linear Regression
2. Random Forest
3. XGBoost

The target variable is:

TotalClaims

```python
X_train, X_test, y_train, y_test = mdl.prepare_data(
    claims_df,
    "TotalClaims"
)

print(X_train.shape)
print(X_test.shape)
```

```python
lr_model, lr_rmse, lr_r2 = mdl.linear_regression_model(
    X_train,
    X_test,
    y_train,
    y_test
)

print("Linear Regression RMSE:", lr_rmse)
print("Linear Regression R2:", lr_r2)
```

```python
rf_model, rf_rmse, rf_r2 = mdl.random_forest_model(
    X_train,
    X_test,
    y_train,
    y_test
)

print("Random Forest RMSE:", rf_rmse)
print("Random Forest R2:", rf_r2)
```

```python
xgb_model, xgb_rmse, xgb_r2 = mdl.xgboost_model(
    X_train,
    X_test,
    y_train,
    y_test
)

print("XGBoost RMSE:", xgb_rmse)
print("XGBoost R2:", xgb_r2)
```

```python
results = pd.DataFrame({

    "Model": [
        "Linear Regression",
        "Random Forest",
        "XGBoost"
    ],

    "RMSE": [
        lr_rmse,
        rf_rmse,
        xgb_rmse
    ],

    "R2": [
        lr_r2,
        rf_r2,
        xgb_r2
    ]
})

results
```

```python
explainer = shap.Explainer(rf_model)

shap_values = explainer(X_test)

shap.plots.beeswarm(shap_values)
```

# Business Interpretation

The Random Forest and XGBoost models outperform Linear Regression, indicating that nonlinear relationships exist between policy characteristics and insurance claims.

Feature importance analysis highlights the variables most associated with elevated claim severity, supporting the development of a more accurate risk-based pricing framework.
